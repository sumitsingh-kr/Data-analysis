```python
data = pd.read_csv("RELIANCE.NS.csv")
```

```python
data
```

```python
#to find the mean or average of a column  
data['Low'].mean()
```

```python
month = pd.to_datetime(data['Date'])
month = month.dt.month_name(locale  = 'English')
```

```python
month
```

```python
type(month)
```

```python
updated_data = month.to_frame()
```

```python
updated_data
```

```python
updated_data.rename(columns={'Date':'Month'}, inplace=True)
```

```python
updated_data
```

```python
data1 = data.join(updated_data)
```

```python
data1
```

```python
data1.groupby(month).mean()
```

```python
avg_stock_price = (data1['High'] + data1['Low'])/2
data1['avg_price'] = avg_stock_price
```

```python
avg_stock_price
```

```python
stock_price_month_wise = data1[['month','avg_price']]
```

```python
stock_price_month_wise
```

```python
type(stock_price_month_wise)
```

```python
plt.figure(figsize=(35,20))

plt.bar(
stock_price_month_wise['month'],
stock_price_month_wise['avg_price'], 
color = (0.5,0.1,0.5,0.6)) 

plt.suptitle('Reliance Stock price over the month', fontsize=20) 
plt.title('Using Data from Years 2016-2021',fontsize=20)  
plt.xlabel('Month', fontsize=20) 
plt.ylabel('Average stock price',fontsize=20)
    
plt.setp(plt.gca().get_xticklabels(),
    rotation=45,
    horizontalalignment='right',
    fontsize=20)
plt.setp(plt.gca().get_yticklabels(), fontsize=20)
```
