```python
reliance_stock_price_month = data1[['Month','avg_price']]
```

```python
reliance_stock_price_month = reliance_stock_price_month.groupby('Month')['avg_price'].sum()
```

```python
reliance_stock_price_month.reset_index(level=0, inplace=True)
```

```python
reliance_stock_price_month
```

```python
plt.figure(figsize=(35,20))

plt.bar(
reliance_stock_price_month['Month'],
reliance_stock_price_month['avg_price'], 
color = (0.5,0.1,0.5,0.6)) 

plt.suptitle('Reliance Stock price over the month', fontsize=20) 
plt.title('Using Data from Years 2016-2021',fontsize=20)  
plt.xlabel('Month', fontsize=20) 
plt.ylabel('Average stock price',fontsize=20)

for i, num in enumerate(reliance_stock_price_month['avg_price']):
    plt.text(
        i,
        num + 10000,
        num,
        ha='center',
        fontsize=15) 
    
plt.setp(plt.gca().get_xticklabels(),
    rotation=45,
    horizontalalignment='right',
    fontsize=20)
plt.setp(plt.gca().get_yticklabels(), fontsize=20)
```
